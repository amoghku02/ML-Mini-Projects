```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset/train.csv')
```

```python
df.head()
```

```python
df.drop('Lead_name', axis=1, inplace=True)
df.drop('Deal_title', axis=1, inplace=True)
df.drop('Contact_no', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Deal_value'].fillna('247839$', inplace=True)
```

```python
df['Industry'].fillna('Banks', inplace=True)
df['Weighted_amount'].fillna('1452260.0168$', inplace=True)
```

```python
df['Weighted_amount'] = df['Weighted_amount'].apply(lambda a: np.float64(a.split('$')[0])).mean()
```

```python
df['Location'].fillna('Unknown', inplace=True)
```

```python
df['Last_lead_update'].fillna('Not Known', inplace=True)
```

```python
df['Resource'].value_counts().sort_values(ascending=True).head(10)
```

```python
df['Resource'].fillna('Not known', inplace=True)
```

```python
def fill_geography(location):
    if(location.find(',')!=-1):
        return 'USA'
    else:
        return 'India'
    
df['Geography'] = df['Location'].apply(fill_geography)
```

```python
df.drop('POC_name', axis=1, inplace=True)
```

```python
df.head()
```

```python
df['Deal_value'] = df['Deal_value'].apply(lambda a: np.int64(a.split('$')[0]))
```

```python
df['Pitch'] = df['Pitch'].apply(lambda a: np.int64(a.split('_')[1]))
df['Fund_category'] = df['Fund_category'].apply(lambda a: np.int64(a.split(' ')[1]))
df.drop(['Location', 'Lead_POC_email'], axis=1, inplace=True)
df['Level_of_meeting'] = df['Level_of_meeting'].apply(lambda a: np.int64(a.split(' ')[1]))
```

```python
def impute_lead_revenue(revenue):
    temp = revenue.split('-')
    value1 = temp[0].split(' ')[0]
    if(temp[1].split(' ')[2] == 'Billion'):
        value2 = np.int64(temp[1].split(' ')[1])*1000
    else:
        value2 = temp[1].split(' ')[1]
    
    return np.int64(value2)-np.int64(value1)

df['Lead_revenue'] = df['Lead_revenue'].apply(impute_lead_revenue)
```
