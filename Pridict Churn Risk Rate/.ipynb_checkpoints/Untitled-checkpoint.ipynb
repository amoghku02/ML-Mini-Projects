```python
def impute_last_visit_time(time):
    ls = time.split(':')
    new_time = np.int64(ls[0]) + (np.int(ls[1])/60) + (np.int(ls[2])/3600)
    return new_time

df['last_visit_time'] = df['last_visit_time'].apply(impute_last_visit_time)
```

```python
def impute_avg_freq_login_days(days):
    if days == 'Error':
        return 0
    else:
        return np.float64(days)
```

```python
df['avg_frequency_login_days'] = df['avg_frequency_login_days'].apply(impute_avg_freq_login_days)
```

```python
df.corr()
```

```python
df['region_category'].fillna('None', inplace=True)
```

```python
df_clean = pd.get_dummies(df, drop_first=True)
```

```python
Y = df_clean['churn_risk_score']
X = df_clean.drop('churn_risk_score', axis=1, inplace=False)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier()
```

```python
forest_clf.fit(X_train, Y_train)
```

```python
y_pred = forest_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(Y_test, y_pred)
```

```python
def data_cleaning(df):
    df.drop('Name', axis=1, inplace=True)
    df.drop('security_no', axis=1, inplace=True)
    df.drop('customer_id', axis=1, inplace=True)
    df['preferred_offer_types'].fillna('None', inplace=True)
    df['points_in_wallet'].fillna(0, inplace=True)   
    df.drop('joining_date', axis=1, inplace=True)
    df.drop('referral_id', axis=1, inplace=True)
    df['points_in_wallet'].fillna(0, inplace=True)
    df['region_category'].fillna('None', inplace=True)
    df['preferred_offer_types'].fillna('None', inplace=True)
    df['last_visit_time'] = df['last_visit_time'].apply(impute_last_visit_time)
    df['avg_frequency_login_days'] = df['avg_frequency_login_days'].apply(impute_avg_freq_login_days)
    df_clean = pd.get_dummies(df, drop_first=True)

    return df_clean
```

```python
test_df = pd.read_csv('dataset/test.csv')
```

```python
test_df_clean = data_cleaning(test_df)
```

```python
test_pred = forest_clf.predict(test_df_clean)
```

```python
n_estimators = [int(x) for x in np.linspace(start=100, stop=1200, num=12)]
#no. of features to consider at every split
max_features = ['auto', 'sqrt']
#maximum no. of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num=6)]
#min no. of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
#min no. of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]


#create random grid
random_grid = {
    'n_estimators':n_estimators,
    'max_features':max_features,
    'max_depth':max_depth,
    'min_samples_split':min_samples_split,
    'min_samples_leaf':min_samples_leaf,
    'bootstrap':[True, False]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
rf = RandomForestClassifier()
```

```python
random_rf = RandomizedSearchCV(estimator=rf, param_distributions=random_grid, scoring='f1_macro', random_state=42, cv=5, verbose=2, n_jobs=-1)
```
